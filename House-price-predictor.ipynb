```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return(float(x))
    except:
        return None
```

**Outlier Detection and Removal**

```python
data.describe()
#here we can see that 1 sqft flat is an outlier as it is not possible
```

```python
#a 1bhk room must be 300 sqft at minimum
data = data[((data['total_sqft']/data['bhk']) >= 300)]
data.describe()
```

```python
#for every location, the mean and standard deviation is calculated 
#only values that fall into m-std to m+std range is kept, rest are discarded
def remove_outlier_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby('location'):
#group by creates buckets of location and their dataframe
        m = np.mean(subdf.price_per_sqft)
        
        std = np.std(subdf.price_per_sqft)
        
        gen_df = subdf[(subdf.price_per_sqft > (m-std)) & (subdf.price_per_sqft <= (m+std))]
        
        df_output = pd.concat([gen_df,df_output],ignore_index=True)
        
    return df_output
data = remove_outlier_sqft(data)
data.describe()
```

```python
data.head()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean':np.mean(bhk_df.price_per_sqft),
                'std' :np.std(bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5:
                #index.values here means fetch index value wherever mean of price_per_sqfr of n-1 bhk flat is greater than price_per_sqft of n bhk flat
                exclude_indices = np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft < stats['mean']].index.values)
    return df.drop(exclude_indices,axis='index')
                
```

```python
data = bhk_outlier_remover(data)
data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
from sklearn.compose import make_column_transformer
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False),['location']),
    remainder = 'passthrough',
)
```

```python
column_trans
```

```python
scaler = StandardScaler()
```

**Linear Regression**

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

**Lasso**

```python
ls = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,ls)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ls = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_ls)
```

**Ridge**

```python
ridge = Ridge()
```

```python
pipe =  make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_ridge)
```

```python
print('No Regularization:', r2_score(y_test,y_pred_lr))
print('Lasso Regression :', r2_score(y_test,y_pred_ls))
print('Ridge Regression :', r2_score(y_test,y_pred_ridge))
```
